# Explolatory and Data Analysis


#### Let's try to get a picture of what kind of roles are typically posted on Naukri.com. We will first look at job experience ranges.




<a id='d'></a>
### Experience


```python
x = df["Job Experience Required"].value_counts().nlargest(10)
plt.figure(figsize=(7,7))
sns.barplot(x=x,y=x.index,data=df)
plt.title(' Top 10 experience ranges')
plt.xlabel('count')
plt.ylabel('Experiences')
```

<a id='e'></a>
### Role category

```python
plt.figure(figsize=(10,10))
role = df['Role Category'].value_counts().nlargest(n=10)
role.plot(kind='pie',autopct='%1.1f%%',explode=(0,0,0,0,0.3,0.30,0,0,0.1,0))

plt.title('Top 10 Role Categories')
```

<a id='f'></a>
### Location

```python
location = df['Location'].value_counts().nlargest(n=10)
plt.figure(figsize=(10,8))
sns.barplot(location.index,location,data=df)
plt.title('Top 10 cities by Job count')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

<a id="g"></a>
### Job Title


```python
title = df['Job Title'].value_counts().nlargest(n=10)
plt.figure(figsize=(9,6))
sns.barplot(title,title.index,data=df)
plt.title('"Top 10  job titles"')
plt.xlabel('Count')
```
